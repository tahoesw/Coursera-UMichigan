## Estimating joint probabilities



If earthquakes on faults are independent and uncorrelated:

$p_{AT}$ := probability of event occuring in time window on one fault (AT)

$q_{AT} = 1 - p_{AT} $ := probability of event not occuring in time window on one fault (AT)

$p_{AT \,or\,LP} = 1 - (q_{AT} \cdot q_{LP})$

```python
fault_list = [ks, lp, gm, slr, nlr, pqxq, pqxn, nqtl, at]

inv_list = [1-f for f in fault_list]
```

```python
all_prob = 1 - np.product(inv_list)
```

```python
all_prob.head()
```

```python
figsize(12,8)

for M in all_prob.columns:
    plot(all_prob.index, all_prob[M], label=M)
legend(loc='upper left', fontsize=16, title='M')
xlabel('observation time window (yr)', fontsize=16)
ylabel('prob. of observing eq >= M',fontsize=16)
title('likelihood of observing a significant earthquake in a time window\n'
      + 'on any of 9 LANFs Richard has compiled', fontsize=18)
```
