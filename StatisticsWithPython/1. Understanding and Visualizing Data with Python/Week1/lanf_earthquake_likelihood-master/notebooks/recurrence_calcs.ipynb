```python
f_stats = {}

for i, g_ in enumerate(g_list):
    g_bin_cols = list( g_.columns)[2:]
    g_bins = np.float_( g_bin_cols)

    g_65 = g_[g_.M == M_min].sum()
    
    #print(faults[i], g_65[g_bin_cols].values.max())
    g_65_counts = g_65[g_bin_cols].values
    
    print(faults[i], g_bin_cols[np.argmax(g_65_counts)])
    
    #g_stats[faults[i]] = dict( zip( g_bins, g_65_counts / 174824.0))
    f_stats[faults[i]] = {'g_bins' : g_bins, 'g_density' : g_65_counts / 174824.0}
    
    
                
    
    #f_max[faults[i]] = g_6.max
```

```python
#c_stats = {}

for i, c_ in enumerate(c_list):
    c_bin_cols = list( g_.columns)[2:]
    c_bins = np.float_( g_bin_cols)

    c_65 = c_[c_.M == M_min].sum()
    
    #print(faults[i], c_65[c_bin_cols].values.max())
    c_65_counts = c_65[c_bin_cols].values
    print(faults[i], g_bin_cols[np.argmax(c_65_counts)])
    
    f_stats[faults[i]]['c_bins'] =  c_bins
    f_stats[faults[i]]['c_density'] =  c_65_counts / 1894874.0
```

```python
figure()

subplot(2,1,1)

for fault in faults:
    
    
    #dips = np.ones(49) * f.loc[fault, 'dip_deg'] 
    dips =  np.random.uniform( f.loc[fault, 'dip_deg'] - f.loc[fault, 'dip_err_deg'],
                               f.loc[fault, 'dip_deg'] + f.loc[fault, 'dip_err_deg'],
                               49)
    
    scatter(dips, f_stats[fault]['g_bins'], c=f_stats[fault]['g_density'],
            vmin = 0, vmax = 1, lw=0, alpha=0.5, cmap='binary')

colorbar()

gca().set_yscale('log')
#title('GR dist')
#xlabel('fault dip')
ylabel('recurrence interval')


subplot(2,1,2)

for fault in faults:
    
    #dips = np.ones(49) * f.loc[fault, 'dip_deg']
    
    dips = np.random.uniform( f.loc[fault, 'dip_deg'] - f.loc[fault, 'dip_err_deg'],
                                 f.loc[fault, 'dip_deg'] + f.loc[fault, 'dip_err_deg'],
                                 49)   
    
    scatter(dips, f_stats[fault]['c_bins'], c=f_stats[fault]['c_density'],
            vmin = 0, vmax = 1, lw=0, alpha=0.5, cmap='binary')

colorbar()

gca().set_yscale('log')
#title('char dist')
xlabel('fault dip')
ylabel('recurrence interval')

#savefig('/Users/itchy/Desktop/rec_int_vs_dip.pdf')
```

```python
figure()

subplot(2,1,1)

for fault in faults:
    
    slip_rates = np.ones(49) * f.loc[fault, 'slip_rate_mm_a'] 
    
    scatter(slip_rates, f_stats[fault]['g_bins'], c=f_stats[fault]['g_density'],
            vmin = 0, vmax = 1, lw=0, alpha=0.5, cmap='binary')

colorbar()

gca().set_yscale('log')
#title('GR dist')
#xlabel('fault dip')
ylabel('recurrence interval')


subplot(2,1,2)

for fault in faults:
    
    slip_rates = np.ones(49) * f.loc[fault, 'slip_rate_mm_a']

    scatter(slip_rates, f_stats[fault]['c_bins'], c=f_stats[fault]['c_density'],
            vmin = 0, vmax = 1, lw=0, alpha=0.5, cmap='binary')

colorbar()

gca().set_yscale('log')
#title('char dist')
xlabel('mean slip rate')
ylabel('recurrence interval')
```
