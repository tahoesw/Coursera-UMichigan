## So what *is* the likelihood of LANF activity?

$ P(A) = 0.5 $ := prior for LANF activity

$ P(\sim A) = 1 - P(A) = 0.5 $ := prior for LANF inactivity

$ P(O \, |\, A) = 0.5 $  := probability of observing LANF quake given activity (this is the result of our study)

$ P(\sim O \,|\,  A) = 1 - P(O \, |\, A) = 0.5 $ := probability of *not* observing LANF earthquake given its activity (also result of this study)

$ P(O \, | \sim A) = 0 $ := probability of false-positive LANF observation

$ P( \sim O | \sim A) = 1 $ := probability of not observing a LANF event given their inactivity


$ P(A | \sim O) = \frac{ P(\sim O|A) P(A)}{P(\sim O|A)P(A) + P(\sim O|\sim A)P(\sim A)} $



```python
%pylab inline
%config InlineBackend.figure_format = 'svg'
```

```python
import numpy as np
#from pylab import *
```

```python
'''priors for lanf activity'''
p_A = np.linspace(0, 1)
```

```python
def no_obs(prior, false_pos, accur):
    p_A = prior
    p_nA = 1 - p_A
    
    p_OgnA = false_pos
    p_nOgnA = 1 - p_OgnA
    
    p_OgA = accur
    p_nOgA = 1 - p_OgA
    
    return p_nOgA * p_A / ( p_nOgA * p_A + p_nOgnA * p_nA)
    
```

```python
plot(p_A, p_A, 'k--')

plot(p_A, no_obs(p_A, 0.01, 0.02), 'g', label='P(O|A) = 0.1')
plot(p_A, no_obs(p_A, 0.01, 0.5), 'b', label='P(O|A) = 0.5')
plot(p_A, no_obs(p_A, 0.01, 0.8), 'r', label='P(O|A) = 0.8')

legend(loc='upper left')


xlabel('prior likelihood for LANF seismicity, P(A)')
ylabel('posterior likelihood, P(A|~O)')

show()
```
